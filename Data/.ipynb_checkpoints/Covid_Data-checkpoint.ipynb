```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
covid_file = "COVID-19_Cases_US.csv"

covid_df = pd.read_csv(covid_file)
```

```python
covid_df
```

```python
covid_state = covid_df.groupby('Province_State')
```

```python
covid_state_confirmed = covid_state['Confirmed'].sum()
```

```python
covid_state_deaths = covid_state['Deaths'].sum()
```

```python
state_confirmed = pd.DataFrame(covid_state_confirmed)
state_deaths = pd.DataFrame(covid_state_deaths)
```

```python
state_merge = state_confirmed.merge(state_deaths, on='Province_State')
```

```python
state_merge['Percent'] = state_merge['Deaths'] / state_merge['Confirmed'] * 100
```

```python
state_merge = state_merge.reset_index()
```

```python
state_merge= state_merge.rename(columns={
    'Province_State':'state',
    'Confirmed':'confirmed',
    'Deaths':'deaths',
    'Percent':'percent'})
```

```python
state_merge
```

```python
state_merge.to_csv('Covid_C_D_byState.csv')
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL_Proj"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
state_merge.to_sql(name='confirmed', con=engine, if_exists='append', index=False)
```
