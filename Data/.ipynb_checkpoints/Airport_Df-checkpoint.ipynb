```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file="COVID-19_Cases_US.csv"
c_df=pd.read_csv(csv_file)
c_df.head()
```

```python
covid_df= c_df[['Last_Update','Confirmed','Deaths','Active','Province_State','Combined_Key','Incident_Rate','FIPS']].copy()
covid_df.head()
```

```python
airport_csv="Airports.csv"
a_df=pd.read_csv(airport_csv, encoding= 'unicode_escape')
a_df.head()
```

```python
us_airport_df= a_df.loc[a_df['country']== 'United States']
us_airport_df
```

```python
s_df= us_airport_df['location'].str.split(',', expand=True)
```

```python
ss_df=s_df.rename(columns={0:'City',1:'State',2:'NA'})
ss_df
```

```python
us_airport_df['State']=ss_df['State']
```

```python
us_airport_df
```

```python
airport_df= us_airport_df.loc[us_airport_df['year']== 2016]
```

```python
airport_df['total\r\npassengers']=airport_df['total\r\npassengers'].str.replace(',','')
```

```python
airport_df
```

```python
airport_count= pd.DataFrame(airport_df['total\r\npassengers'].astype(str).astype(int))
```

```python
airport_count
```

```python
airport_df=airport_df.drop(columns=['total\r\npassengers'])
```

```python
airport_df['Passenger Count']=airport_count['total\r\npassengers']
```

```python
airport_df
```

```python
count_df= airport_df.groupby('State')
passenger_count=count_df['Passenger Count'].sum()
```

```python
airport_passenger_df=pd.DataFrame(passenger_count)
```

```python
airport_passenger_df
```

```python
covid_airport_df=airport_df[['airport','State','Passenger Count','year']]
```

```python
covid_airport=covid_airport_df.rename(columns={
    'State':'state',
    'Passenger Count':'passenger_count'})
```

```python
covid_airport
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL_Proj"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
covid_airport.to_sql(name='airport', con=engine, if_exists='append', index=False)
```
