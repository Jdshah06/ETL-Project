```python
import pandas as pd
import numpy as np
```

```python
covid_file = "COVID-19_Cases_US.csv"

covid_df = pd.read_csv(covid_file)
```

```python
covid_df
```

```python
covid_state = covid_df.groupby('Province_State')
```

```python
covid_state_confirmed = covid_state['Confirmed'].sum()
```

```python
state_confirmed = pd.DataFrame(covid_state_confirmed)
```

```python
state_confirmed
```

```python
state_confirmed.to_csv('Covid_Confirmed_byState')
```

```python
covid_state_deaths = covid_state['Deaths'].sum()
```

```python
state_deaths = pd.DataFrame(covid_state_deaths)
```

```python
state_deaths
```

```python
state_merge = state_confirmed.merge(state_deaths, on='Province_State')
```

```python
state_merge['Percent'] = state_merge['Deaths'] / state_merge['Confirmed']
```

```python
state_new = state_merge.style.format({'Percent':'{:,.2%}'})
```

```python
state_new
```

```python
state_new.to_excel('Covid_C_D_byState')
```
