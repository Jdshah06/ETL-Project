```python
import pandas as pd
import numpy as np
```

```python
covid_file = "COVID-19_Cases_US.csv"

covid_df = pd.read_csv(covid_file)
```

```python
covid_df
```

```python
covid_state = covid_df.groupby('Province_State')
```

```python
covid_state_confirmed = covid_state['Confirmed'].sum()
```

```python
state_confirmed = pd.DataFrame(covid_state_confirmed)
```

```python
state_confirmed
```

```python
state_confirmed.to_csv('Covid_Confirmed_byState')
```
